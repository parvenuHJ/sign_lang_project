### CNN 의 발전
- ResNet (Residual Network):  
특징: 잔여 학습을 도입하여 네트워크 깊이를 키우면서 그레이디언트 소실 문제를 완화한 ResNet은 이미지 특징 추출에서 널리 사용됩니다.  
- DenseNet:  
특징: 밀집 연결(Dense Connectivity)을 사용하여 각 레이어가 이전 레이어의 출력에 직접 연결되는 구조를 가진 DenseNet은 특징 추출 및 이미지 분류에 사용됩니다.

위 모델들 설명 : https://warm-uk.tistory.com/46

- Xception:  
특징: InceptionV3의 확장으로, 깊은 네트워크를 구축하면서도 파라미터의 수를 줄이는 Xception은 이미지 특징 추출에서 사용됩니다.
Xception 모델을 사용하여 딥러닝을 수행하려면 다음 단계를 따르면 됩니다. 주로 TensorFlow와 Keras를 사용하는 것을 가정하고, 필요에 따라 환경에 맞게 수정할 수 있습니다.
  
논문 출처 : https://koreascience.kr/article/JAKO201926358474432.pdf (CNN의 깊은 특징과 전이학습을 사용한 보행자 분류)
그런데 이미지 크기가 기본적으로 299x299 를 주로 사용한다고함.
